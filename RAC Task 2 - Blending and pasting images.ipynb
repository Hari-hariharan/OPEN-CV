<h1>2.Blending and Pasting images</h1>

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img1=cv2.imread("ship.jpg")
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
```

```python
img2=cv2.imread("nature.jpg")
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
img1=cv2.resize(img1,(600,600))
```

```python
plt.imshow(img2)
```

```python
img2.shape
```

```python
x_offset=1000-600
y_offset=1250-600
```

```python
img1.shape
```

```python
rows,cols,channels=img1.shape
```

```python
roi=img2[y_offset:1250,x_offset:1000]
```

```python
img1_gray=cv2.cvtColor(img1,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(img1_gray,cmap="gray")
```

```python
mask_inv=cv2.bitwise_not(img1_gray)
```

```python
plt.imshow(mask_inv,cmap="gray")
```

```python
mask_inv.shape
```

```python
white_background = np.full(img1.shape,255,dtype=np.uint8)
```

```python
white_background.shape
```

```python
bk = cv2.bitwise_or(src1=white_background,src2=white_background,mask=mask_inv)
```

```python
bk.shape
```

```python
fg = cv2.bitwise_or(img1,img1,mask=mask_inv)
```

```python
final_roi=cv2.bitwise_or(roi,fg)
```

```python
plt.imshow(final_roi)
```

```python
large_img=img2
small_img=final_roi
```

```python
large_img[y_offset:y_offset+small_img.shape[0],x_offset:x_offset+small_img.shape[1]]=small_img
```

```python
plt.imshow(large_img)
```
